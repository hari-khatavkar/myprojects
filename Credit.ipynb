Since the plan is mostly to use a trial and error approach to see how things go, I will first start on those that I am more familiar with: Decision tree

```python
#importing libraries

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, make_scorer
from sklearn.model_selection import train_test_split
```

```python
features = ['LIMIT_BAL', 'SEX', 'EDUCATION', 'MARRIAGE', 'AGE', 'PAY_1', 'PAY_2', 'PAY_3', 'PAY_4', 'PAY_5', 'PAY_6', 'BILL_AMT1', 'BILL_AMT2', 'BILL_AMT3', 'BILL_AMT4', 'BILL_AMT5', 'BILL_AMT6', 'PAY_AMT1', 'PAY_AMT2', 'PAY_AMT3', 'PAY_AMT4', 'PAY_AMT5', 'PAY_AMT6']
X = df[df.columns[:-1]].copy()
y = df['Y'].copy()
```

```python
#Train - Test Split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42) # this means we will train on 75% of the data and test on the remaining 25%.
```

```python
# creating an instance of the decision tree classifier
classifier = DecisionTreeClassifier(max_depth=10, random_state=42)
# max_depth=10 - The maximum depth of the tree.
# random_state=42 - Controls the randomness of the estimator. 

# training the classifier
%time classifier.fit(X_train, y_train)
# do our predictions on the test
predictions = classifier.predict(X_test)
# see how good we did on the test
accuracy_score(y_true = y_test, y_pred = predictions)
```

 - 81% accuracy is a result we just get with a random choices on the parameters. Not bad for a beginner.
 - On this matter, let's see an overfitting example by changing the max_depth value to 100

```python
classifier = DecisionTreeClassifier(max_depth=100, random_state=14) 
%time classifier.fit(X_train, y_train)
predictions = classifier.predict(X_test)
accuracy_score(y_true = y_test, y_pred = predictions)
```

- Hence, model accuracy decreased

```python
#A better metric is the f1-score, which takes into account the false positives, the false negatives etc.
#F1-score is the harmonic mean of recall and precsiion 

from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import KFold
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
from sklearn import metrics
from sklearn.metrics import confusion_matrix

#Performing GridSearchCV to find the best parameters

param_grid = {'max_depth': np.arange(3, 10),
             'criterion' : ['gini','entropy'],
             'max_leaf_nodes': [5,10,20,100],
             'min_samples_split': [2, 5, 10, 20],
             'class_weight' : ['balanced']} #This is the additional paramenter 
grid_tree = GridSearchCV(DecisionTreeClassifier(), param_grid, cv = 5, scoring= 'f1')
%time grid_tree.fit(X_train, y_train)
best = grid_tree.best_estimator_
print(grid_tree.best_estimator_)
print("Best Recall Score: ",np.abs(grid_tree.best_score_))

#Fitting model with best parameters obtained from GridSearchCV operation

classifier = best
%time classifier.fit(X_train, y_train)
predictions = classifier.predict(X_test)
print("-------------")
print(f1_score(y_true = y_test, y_pred = predictions))
print("-------------")

#Confusion Matrix

TN, FP, FN, TP = confusion_matrix(y_test, predictions).ravel()
print("True Positive (TP):", TP)
print("True Negative (TN):", TN)
print("False Positive (FP):", FP)
print("False Negative (FN):", FN)

#Classification report

print(classification_report(y_test, predictions))
```

```python
#Feature Importance

def get_feature_importance(clsf, ftrs):
    imp = clsf.feature_importances_.tolist()
    feat = ftrs
    result = pd.DataFrame({'feat':feat,'score':imp})
    result = result.sort_values(by=['score'],ascending=False)
    return result

%time get_feature_importance(classifier, features)
```

- Now that we have completed our blind machine learning, lets head on to do some data visualization to understand our data in a better manner and gather some underlying insights.
- Also, it can be noted that, "Pay_1" has significantly ranked the highest in feature importance.

```python
#Now buidling our model just on that one feature just for our anaoysis and some fun!

X = df[['PAY_1']].copy()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

%time classifier.fit(X_train, y_train) #same classifier as obtained from GridSearchCV
predictions = classifier.predict(X_test)
print(accuracy_score(y_true = y_test, y_pred = predictions))

TN, FP, FN, TP = confusion_matrix(y_test, predictions).ravel()
print("True Positive (TP):", TP)
print("True Negative (TN):", TN)
print("False Positive (FP):", FP)
print("False Negative (FN):", FN)

print(classification_report(y_test, predictions))
```

```python
#Running all possible classification models in pipeline fashion and storing their accuracy and f-score

models = []

#Instanctiating the models

models.append(('LR', LogisticRegression(max_iter=1000000)))
models.append(('LDA', LinearDiscriminantAnalysis()))
models.append(('KNN', KNeighborsClassifier(n_neighbors=20)))
models.append(('CART', DecisionTreeClassifier()))
models.append(('NB', GaussianNB()))
models.append(('SVM', SVC()))
models.append(('XGB', XGBClassifier()))
models.append(('ADA', ensemble.AdaBoostClassifier()))
models.append(('BAG', ensemble.BaggingClassifier()))
models.append(('TRE', ensemble.ExtraTreesClassifier()))
models.append(('GRA', ensemble.GradientBoostingClassifier()))
models.append(('RAN', ensemble.RandomForestClassifier()))
models.append(('PAS', linear_model.PassiveAggressiveClassifier()))
models.append(('RID', linear_model.RidgeClassifierCV()))
models.append(('SDG', linear_model.SGDClassifier()))
models.append(('PER', linear_model.Perceptron()))
#Navies Bayes
models.append(('BER', naive_bayes.BernoulliNB()))
models.append(('GAU', naive_bayes.GaussianNB())) 
#SVM
#models.append(('NUS', svm.NuSVC(probability=True)))
models.append(('LIN', svm.LinearSVC()))
#Trees    
models.append(('EXT', tree.ExtraTreeClassifier()))
models.append(('QDA', discriminant_analysis.QuadraticDiscriminantAnalysis()))

results = []
results_2 = []
results_3 = []
names = []

#Running all models with k-fold cross validation
from sklearn.model_selection import cross_val_predict
dwPreds = []
for name, model in models:
    # store the metrics
    %time cv_results = cross_val_score(model, X_train, y_train, cv = 5, scoring='accuracy')
    %time cv_results_2 = cross_val_score(model, X_train, y_train, cv = 5, scoring='f1')
    %time cv_results_3 = cross_val_score(model, X_train, y_train, cv = 5, scoring='recall')
    results.append(cv_results)
    results_2.append(cv_results_2)
    results_3.append(cv_results_3)
    names.append(name)
    msg = "%s: Acc - %f (%f), f1 - %f (%f), Rec - %f (%f)" % (name, cv_results.mean(), cv_results.std(), cv_results_2.mean(), cv_results_2.std(), cv_results_3.mean(), cv_results_3.std())
    print(msg)
```
